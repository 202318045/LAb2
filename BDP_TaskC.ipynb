```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://www.amazon.in/s?k=nerf&ref=nb_sb_noss_2'

HEADERS = ({'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36','Accept-Language' : 'en-US, en ;q = 0.5'})
```

```python
webcode = requests.get(URL, headers = HEADERS)
```

```python
webcode
```

```python
soup = BeautifulSoup(webcode.content, "html.parser")
soup
```

```python
links  = soup.find_all("a", attrs = {'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
link = links[0].get('href')
link
```

```python
product_list = 'https://amazon.in' + link
product_list
```

```python
webpage = requests.get(product_list, headers = HEADERS)
webpage
```

```python
soup1 = BeautifulSoup(webpage.content, "html.parser")
soup1
```

```python
soup1.find("span", attrs = {"id" : "productTitle"}).text.strip()
```

```python
soup1.find("span", attrs = {"class" : "a-price-whole"}).text
```

```python
soup1.find("span", attrs = {"class" : "a-icon-alt"}).text
```

```python
def get_title(soup):
    title_element = soup.find("span", attrs={"id": "productTitle"})
    if title_element:
        title = title_element.text.strip()
        return title
    else:
        return None

def get_price(soup):
    price_element = soup.find("span", attrs={"class": "a-price-whole"})
    if price_element:
        price = price_element.text
        return price
    else:
        return None

def get_rating(soup):
    rating_element = soup.find("span", attrs={"class": "a-icon-alt"})
    if rating_element:
        rating = rating_element.text
        return rating
    else:
        return None
```

```python
links_list = []
for link in links:
  links_list.append(link.get('href'))

D = {'title':[], 'price':[], 'rating':[]}

for link in links_list:
  webpage1 = requests.get('https://amazon.in/' + link, headers = HEADERS)
  soup2 = BeautifulSoup(webpage1.content, "html.parser")

  D['title'].append(get_title(soup2))
  D['price'].append(get_price(soup2))
  D['rating'].append(get_rating(soup2))

amazon_data = pd.DataFrame.from_dict(D)
amazon_data.to_csv("amazon_data.csv", header = True, index = False)
```
